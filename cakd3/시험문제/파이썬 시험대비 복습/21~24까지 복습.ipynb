```python
# Q. 반복문으로 팩토리얼을 구하는 사용자 함수를 작성하고 5!, 10!를 구하세요.

def re1(a):
    sum = 1
    for i in range(1, 1+a):
        sum *=i
    return sum 
print(re1(5))
print(re1(10))
```

```python
#Q. 재귀란 '자기자신을 호출하는 것'을 의미한다. 재귀함수를 이용해서 팩토리얼을 구하는
#작성하고 1!, 5!, 10!를 구하세요.
#factorial(n) = n * factorial(n-1) (n>=1)

def fac1(a):
    if a == 1 :
        return 1
    elif a > 1:
        return a*fac1(a-1)
    
print(fac1(10))
```

```python
#Q. 랜덤하게 1000명의 이름, 키와 몸무게 만들기

import random
hanguls = list('가나다라마바사아자차카타파하')
with open('info.text', 'w') as f:
    for i in range(1000):
        name = random.choice(hanguls)+random.choice(hanguls)
        height = random.randrange(100,200)
        weight = random.randrange(40, 90)
        f.write("{},{},{}\n".format(name,height,weight))   
```

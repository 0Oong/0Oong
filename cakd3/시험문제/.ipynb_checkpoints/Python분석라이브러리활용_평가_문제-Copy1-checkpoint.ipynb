### Q12.  tdf1의 sex, class 칼럼을 '_'으로 연결한 'sc'칼럼을 추가한 후 아래와 같이 출력하세요.

male_Third

```python
import seaborn as sns
df = sns.load_dataset('titanic')
tdf = df[['survived','sex','age','class']]
tdf.head()

def cat_age(age):
    cat = ''
    if age <= 5: 
        cat = 'Baby'
    elif age <= 12:
        cat = 'Child'
    elif age <= 18: 
        cat = 'Teenager'
    elif age <= 25: 
        cat = 'Student'
    elif age <= 60: 
        cat = 'Adult'
    elif age > 60 : 
        cat = 'Elderly'
    return cat
        
tdf1 = tdf.copy()
# tdf1 = tdf1.age.apply(cat_age)
cat_age = tdf1.age.transform(cat_age)
tdf1['age'] = cat_age
tdf1.head()
```

```python
tdf1['sc'] = tdf[['sex','class']].agg('-'.join, axis=1)
tdf1
```

### Q20.  titanic 데이터셋에서 다음 전처리를 수행하세요.
1. df에서 중복 칼럼으로 고려할 수 있는 컬럼들(6개 내외)을 삭제한 후 나머지 컬럼들로 구성되는 데이터프레임을 df1 이름으로 저장 후 출력하세요. 
2. df1에서 null값이 50% 이상인 칼럼을 삭제 후 df2 이름으로 저장하고 출력하세요.
3. df2에서 결측값이 있는 age 칼럼에 대해서 평균값으로 대체 처리를 수행하세요. 
4. df2에서 결측값이 있는 embarked 칼럼에 대해서 앞행의 값으로 대체 처리를 수행하세요. 
5. df2 문자로 되어있는 칼럼들을 레이블 인코딩 수행하여 숫자로 변환 후 df2.info()를 출력하세요

```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
# df에서 중복 칼럼으로 고려할 수 있는 컬럼들(6개 내외)을 삭제한 후 
# 나머지 컬럼들로 구성되는 데이터프레임을 df1 이름으로 저장 후 출력하세요.
df.columns
df1 = df[['survived','pclass','sex','age','sibsp','parch','fare','embarked','deck']]
df1.head()
```

```python
# df1에서 null값이 50% 이상인 칼럼을 삭제 후 df2 이름으로 저장하고 출력하세요.
df1.isnull().sum()
df2 = df1.drop('deck',axis=1)
df2.head()
```

```python
# df2에서 결측값이 있는 age 칼럼에 대해서 평균값으로 대체 처리를 수행하세요.
df2.age.fillna(df2.age.mean(), inplace=True)
df2.age.isnull().sum()
```

```python
# df2에서 결측값이 있는 embarked 칼럼에 대해서 앞행의 값으로 대체 처리를 수행하세요.
df2.embarked.fillna(df2.embarked.ffill(), inplace=True)
df2.embarked.isnull().sum()
```

```python
# df2 문자로 되어있는 칼럼들을 레이블 인코딩 수행하여 숫자로 변환 후 df2.info()를 출력하세요
from sklearn import preprocessing

features = ['sex', 'embarked']
for feature in features:
    le = preprocessing.LabelEncoder()
    df2[feature] = le.fit_transform(df2[feature])
df2.info()
```

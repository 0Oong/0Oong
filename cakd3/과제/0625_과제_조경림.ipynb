#### Q1. 문자열 리스트 states를 정형화하여 아래와 같이 출력하세요.

states = [['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',  
          'south   carolina##', 'West virginia?']  
[출력]  
['Alabama',  
 'Georgia',  
 'Georgia',  
 'Georgia',  
 'Florida',  
 'South   Carolina',  
 'West Virginia']  

```python
#방법1
states = [' Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
'south carolina##', 'West virginia?']

def a1(character):
    return character.strip(' !#?').title()

list(map(lambda x : a1(x), states))
```

```python
#방법2
states = [' Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
'south carolina##', 'West virginia?']

def a1(character):
    a=character.strip(' !#?')
    b=a.title()
    return b
list(map(lambda x : a1(x), states))
```

```python
#방법3
states = [' Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
'south carolina##', 'West virginia?']

import re

def nor1(strings):
    result = []
    for i in strings:
        i = i.strip()
        i = re.sub('[!#?]', '', i)
        i = i.title()
        result.append(i)
    return result

nor1(states)
```

결과 : 과체중  
BMI가 18.5 이하면 저체중 ／ 18.5 ~ 22.9 사이면 정상 ／ 23.0 ~ 24.9 사이면 과체중 ／ 25.0 이상부터는 비만으로 판정.

```python
#방법 1
import random
hanguls = ('가나다라마바사아자차카타파하')
    
def bmi1(height, weight):
    height = height * 0.01
    bmi = weight / (height*height)
    if bmi <= 18.5 :
        return '결과 : 저체중'
    elif 18.5 < bmi <30.0 :
        return'결과 : 정상'
    elif 23.0 < bmi < 24.9 : 
        return'결과 : 과체중'
    else : 
        return'결과 : 비만'
        
with open('profile.txt', 'w') as file:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        height = random.randrange(140, 200)
        weight = random.randrange(40, 80)
        bmi = weight / (height*height*0.01*0.01)
        file.write("이름 : {}\n키 : {}\n몸무게 : {}\nBmi : {:0.2f}\n{}\n\n".format(name,height,weight,bmi,bmi1(height, weight)))
        
with open("profile.txt",'r') as file:
    data = file.read()
    print(data)
```

#### Q3. 사칙연산 기능을 포함하는 모듈 cal4.py를 작성 후 임포트해서 사칙연산을 수행하세요

```python
def cal4(tp, a, *args):
    if tp == 'add':
        result = a
        for i in args:
            result +=i
            
    elif tp == 'minus':
        result = a
        for i in args:
            result -= i
            
    elif tp == 'multiple':
        result = a 
        for i in args:
            result *= i
            
    elif tp == 'division':
        try :
            result = a
            for i in args:
                result /= i
                
        except ZeroDivisionError:
            print('0은 입력할 수 없습니다. 다시 입력해주세요')
            
    else :
        print('잘못 입력하셨습니다.')    
    print(result)
    
    
cal4('division', 5, 0)
```

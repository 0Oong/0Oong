#### Q11. 게시물의 총 건수와 한 페이지에 보여줄 게시물 수를 입력시 총 페이지 수를 출력하는 프로그램

```python
#방법1
def page1(a,b):
    a = int(input("게시물의 총 건수를 입력하세요: "))
    b = int(input("게시물 수를 입력하세요: "))
    if a % b == 0:
        return a // b 
    elif a / b < 1:
        return 1
page1(5,4)
```

```python
#방법2
def getTotalPage(m,n):
    if m % n == 0:
        return m //  n
    else :
        return m // n + 1
print(getTotalPage(100,3))
```

#### Q11. 랜덤하게 1000명의 이름, 키와 몸무게 만들기
- 이름 : 라바  
- 몸무게 : 80  
- 키 : 146  

```python
# 방법1.
import random
hanguls = list('가나다라마바사아자차카타파하')
with open('info.txt','w') as file:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        height = random.randrange(140,200)
        weight = random.randrange(40,100)
        file.write("{},{},{}\n".format(name,height,weight))
        
with open("info.txt",'r') as file:
    data = file.read()
    print(data)
```

```python
# 방법2.
with open('info.txt','r') as file:
    for line in file:
        (name, height, weight) = line.strip().split(',')
        print('\n'.join([
            "이름:{}",
            "키:{}",
            "몸무게:{}"
            ]).format(name,height,weight))
        print()
```

## name == "main"의미
- 메인 함수의 선언, 시작을 의미합니다. 해당 코드 밑에 main 등의 함수 호출 코드를 작성해서 함수의 기능을 수행
- 직접 mod.py 파일을 실행하는 경우에는 name_ 변수에 main 이라는 값이 할당되고 이 조건절이 참이 되어 if문 다음 문장들이 수행
- 모듈을 불러와서 사용하는 경우 name 변수에는 모듈 이름이 저장되고 이 조건절이 거짓이 되어 if문 다음 문장들이 수행되지 않음
- if name == "main"와 같이 조건문을 사용하여 터미널에서와 같이 직접 호출되어 사용될 때는 그 자체로 기능을 수행하고, 동시에 다른 모듈에서 필요한 함수 등을 제공

예를들어 hello.py이라는 이름으로 간단한 함수를 만들었음

def hello(name):
    return f'Hello~{name}'
print(hello('batman')
결과 : Hello~batman

이 함수가 만약에 다른데도 사용하고 싶어서 다른 파일 index.py를 만들고 hello를 사용하려면 import를 사용해야함

import hello  #우리가 hello라는 모듈을 만들어서 불러온 것임

print(hello.hello('superman'))

결과 : Hello~batman
       Hello~superman
둘 다 나옴
즉, hello.py에 있던 베트맨까지 출력이 다 된거임. 
그런데 우리는 함수만 쓰고 싶은 건데 어떻게 해야할까? 그러면 print(hello('batman')을 지워야할까 싶지만
따로 hello.py를 실행해야할 때도 있기때문에 그럴 수 없음.
그때실행하는게
if__name__ =="__main__":
    print(hello('batman'))
    
그래서 hello.py파일에 넣어줘서
def hello(name):
    return f'Hello~{name}'
if__name__ =="__main__":
print(hello('batman')
해줌

이렇게 넣어주고 index.py를 실행시켜주면
Hello~superman이 실행됨

long_task 함수는 수행하는 데 5초의 시간이 걸리는 함수이다. 위 프로그램은 이 함수를 수행하는 프로그램이다.
이 프로그램은 5초가 5번 반복되니 총 25초의 시간이 걸린다. 

```python
import time
def long_task(): #5s의 시간이 걸리는 함수 
    for i in range(5):
        time.sleep(1) #1초 대기
        print("working:%s\n" %i)
        
print('start')

for i in range(5) : #long_task를 5회 수행
    long_task()
    
print('End')
```

```python
import time
import threading

def long_task(): #5s의 시간이 걸리는 함수 
    for i in range(5):
        time.sleep(1) #1초 대기
        print("working:%s\n" %i)
        
print('start')

threads = []
for i in range(5):
    t = threading.Thread(target=long_task)  #여러가지 일을 동시에 진행시킬 수 있는thread
    threads.append(t)
for t in threads:
    t.start()
for t in threads:
    t.join() #join으로 스레드가 종료할 때까지 기다린다. 이걸 안하면 end가 먼저 나오는 등 뒤죽박죽
print('end')
```

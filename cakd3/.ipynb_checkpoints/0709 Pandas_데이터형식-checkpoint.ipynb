```python
# 딕셔너리로 5행 5열 데이터 프레임을 작성하세요.
import pandas as pd
import numpy as np

index = ['r0','r1','r2','r3','r4']
column = ['a','b','c','d','e']

data = {'a': np.random.randint(1,100, size=5),
        'b': np.random.randint(1,100, size=5),
        'c': np.random.randint(1,100, size=5),
        'd': np.random.randint(1,100, size=5),
        'e': np.random.randint(1,100, size=5)}

df = pd.DataFrame(data, index =['r0','r1','r2','r3','r4'])

df

```

```python
# df에 r5,r6 2개의 행을 추가하고 값은 0을 적용하여 출력하세요.

df.loc['r5'] = 0
df.loc['r6'] =0
df
```

```python
new_index = ['r0','r1','r2','r3','r4','r5','r6','r7','r8']
df5 = df.reindex(new_index, fill_value=0)
df5
```

```python
# r5~r8까지 삭제 후 출력하세요
df1 = df5.drop(['r5','r6','r7','r8'], axis=0)
df1
```

```python
# index 역으로 정렬하고 싶을 때
df1_s = df1.sort_index(ascending=False)
df1_s
```

```python
# 컬럼 정렬
df1_c=df1.sort_values(by='c')   #c를 기준으로 sort
df1_c
```

```python
df1_c2 = df1.sort_values(by='c', ascending=False) #역순으로 기준
df1_c2
```

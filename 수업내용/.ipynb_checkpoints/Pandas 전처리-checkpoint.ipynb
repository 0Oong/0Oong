## 회귀모델 - 종속변수냐 아니냐
우리는 프로그램을 한다는 것은 이런 걸 다 경우의 수를 생각해서 3이나 5를 결정해줌. 데이터가 많으면 많을수록 학습을 충분하게해서 더 정확한 가중치(기울기)를 찾아줄 수 있음. 

회귀모델에 적합한 알고리즘이 있음

## 분류모델 -  0이냐 1이냐
그런데 1차 방정식을 보면 예측값이 연속적인 숫자로 나옴 그런걸 회귀모델이라고 함. 그런데 타이타닉에 각 승객별로 여러가지 조건을 변수화 해서 데이터를 만드는데 조건에 따라 승객이 생존했느냐 사망했느냐를 예측할 수 있을 것. a라는 고객의 조건이 선실1등급,b는 2등급에 타고 나이가 어떻고 이런걸 보면 변수들로 학습해서 이 사람이 살았는지 죽었는지 예측 모델을 만들 수 있음.

분류모델에 적합한 알고리즘이 있음


```python
# Counter
from collections import Counter 
my_list = ['a','b','c','a','b','c','a']
count = Counter(my_list)
print(count)  #b와 c의 갯수가 같지만 b가 먼저라 나옴.
words = dict(count.most_common(n=2)) #가장 많이 나오는 것 두개를 출력하라는 뜻
print(words)  
```

```python
#반올림
import pandas as pd
import numpy as np
col1 = np.random.uniform(10,20,size=10) #10~20사이 실수 10개
col2 = np.random.uniform(10,20,size=10) 

sr1 = pd.Series(col1, name='col1')
sr2 = pd.Series(col2, name='col2')
df = pd.concat([sr1,sr2],axis=1)
df = df.round({'col1':2, 'col2': 1}) #소수점 자리수를 다르게 하고 싶을 때
df.head(3)
```

```python
# none값 만드는 법
df.iloc[1:3,:]=None
df.iloc[1:3,:] = np.nan
df.head()
```

```python
df.fillna(0, inplace=True) #0으로 채우기 fillna는 na를 채운다는 뜻임.
df.info() 
#float타입을 int로 바꾸기
df = df.astype(int)
df.info()
#각각의 타입을 바꾸는 방법
df=df.astype({'col1':'int','col2':'float'})
df.info()
```

```python
# 칼럼 새로 만들기
df['col0']=list('abcdefghij')
df.set_index('col0',inplace=True)
df
```

```python
df1 = df.copy()
df1.loc['c', 'col2']=np.nan
df1
```

```python
# Q. col2의 평균값으로 null값 대체하세요. 대체하는거 -> replace 쓰자
# df3 = df1.replace(np.nan,df1['col2'].mean())
df3 = df1.fillna(df1['col2'].mean())
df3 = df.round({'col2':1})
df3
```

```python
# col2를 정수로 변환하여 출력하세요
# 1) df3=df3.astype({'col2':'int'})
# 2) df3['col2'] = df3['col2'].astype(int)
df3['col2'] = df3.col2.astype(int)
df3.head() 
```

```python
# b를 Nan처리하기
df3.iloc[1:3,1:2]=np.nan
df3.head()
```

```python
# nan있는걸 삭제하기(null값 삭제)
# df3.dropna(axis=0)
df3.dropna(subset=['col2'],inplace=True)
df3.head() 
```

```python
import numpy as np
import pandas as pd
data1 = np.arange(1,21).reshape(4,5)
data2 = np.arange(11,31).reshape(5,4)

df1 = pd.DataFrame(data1, columns=list('abcde'))
df2 = pd.DataFrame(data2, columns=list('dehi'))

print(df1,'\n')
print(df2)
```

```python
con2 = pd.concat([df1,df2],axis=1)
con2.head()
```

```python
# merge() 결합
df1['ind']=[1,2,3,4]
df1

```

```python
df2['ind']=[1,2,3,4,5]
df2
```

```python
pd.merge(df1,df2, on='ind',how='outer')  #공통된건 on이라고 써주면 됨 합집합일떈outer,교집합일땐 inner
# 공통된 애들은 구분돼서 나옴, inner가 디폴트(교집합)
```

```python
df1 = df1.drop('ind',axis=1)
df2 = df2.drop('ind', axis=1)
```

```python
# join은 index를 기준으로 결합, 디폴트가 교집합 how='inner'
df1.join(df2,lsuffix='_1',rsuffix='_2')
```

```python
df1.join(df2,lsuffix='_1',rsuffix='_2',how='outer')
```

```python
df1['ind']=[1,2,3,4]
df2['ind']=[1,2,3,4,5]
```

```python
df1.set_index('ind').join(df2.set_index('ind'),lsuffix='_1',rsuffix='_2', on='ind', how='outer')
```

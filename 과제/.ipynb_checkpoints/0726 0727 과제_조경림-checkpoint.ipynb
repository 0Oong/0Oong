```python
Q1 = df.Start.describe()['25%']
Q3 = df.Start.describe()['75%']
IQR = Q3-Q1

outlier = []
for i in df.Start:
    if not Q1 - IQR*1.5 < i < Q3 + IQR*1.5:
        outlier.append(i)
print(outlier)
```

```python
# [과제]
# Q. df_auto의 각 칼럼을 시각화해서 탐색한 후 인사이트를 기술하세요.(5개 이상)
import pandas as pd
df_auto = pd.read_excel('/Users/joy/cakd3/visualization/dataset/auto-mpg.xlsx')

df_auto
```

```python
#mpg 인덱스로
df_auto.set_index('mpg', inplace=True)
df_auto = df_auto.sort_index()

print(df_auto.info())
df_auto.head(10)
```

```python
# 1 '?'처리
import numpy as np
df_auto['horsepower'] = df_auto['horsepower'].replace('?', np.nan)
df_auto.info()
# null값 채우기
df_auto['horsepower'] = df_auto.horsepower.fillna(df_auto.horsepower.mean())


df_auto.horsepower.plot()

#결론 : 마력이 높을수록 mpg(연비)가 낮다
```

```python
df_auto.reset_index(inplace=True)
df_auto
```

```python
#  3번에서 난 게 더 연비가 좋음 - 비교. 1.미국 2.유럽 3.일본
df_auto.plot(x='origin',y='mpg', kind='scatter')
```

```python
print(df_auto.groupby('origin')['mpg'].count())

df_auto.groupby('origin')['mpg'].mean().plot(kind='bar')
```

```python
# 3 생산연도와 연비
df_auto.groupby('model year')['mpg'].mean().plot()
```

```python
# 4 무게와 연비 - 무게가 무거울수록 연비가 낮다
df_auto.groupby('weight')['mpg'].mean().plot()
```

```python
# 5 가속도와 연비
df_auto.groupby('acceleration')['mpg'].mean().plot()
```

```python
# [과제]
# Q. mpg 컬럼의 이상치를 구하세요.

print(df_auto.mpg.plot(kind='box'))
df_auto.mpg.describe()

#이상치
Q1 = df_auto.mpg.describe()['25%']
Q3 = df_auto.mpg.describe()['75%']
IQR = Q3-Q1
outlier = []
for i in df_auto.mpg:
    if not Q1-IQR*1.5< i < Q3+IQR*1.5:
        outlier.append(i)
print('이상치 : ', outlier)
```

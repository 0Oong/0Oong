```python
# 과제
# Q. offset.rollforward 기준으로 그룹화를 진행 후 그룹별 평균값을 출력하세요.
import numpy as np
import pandas as pd
from pandas.tseries.offsets import Day, MonthEnd, Week, MonthBegin

ts = pd.Series(np.random.randn(20),
               index=pd.date_range('1/1/2000', periods=20, freq='4d'))

# 1 
# 처음 썼던 오답, a = st.groupby('index').offsets.rollforward()
offset1 = MonthEnd()
a1 = ts.groupby(offset1.rollforward).mean()
a2 = ts.groupby(offset1.rollback).mean()
print('<첫번째그룹>')
print(a1)
print(a2,'\n')

# 2
offset2 = Day(3)
b1 = ts.groupby(offset2.rollforward).mean()
b2 = ts.groupby(offset2.rollback).mean()
print('<두번째그룹>')
print(b1)
print(b2,'\n')

# 3
offset3 = Week()
c1 = ts.groupby(offset3.rollforward).mean()
c2 = ts.groupby(offset3.rollback).mean()
print('<세번째그룹>')
print(c1)
print(c2, '\n')

# 4
offset4 = MonthBegin()
d1 = ts.groupby(offset4.rollforward).mean()
d2 = ts.groupby(offset4.rollback).mean()
print('<네번째그룹>')
print(d1)
print(d2, '\n')
```

```python
# 과제
# Q. 상기 문제를 resample을 사용하여 수행하세요. (위에 했던걸 resample로만 바꾸면 됨)
# 1 

a= ts.resample('M').mean()
print('<첫번째그룹>')
print(a)

# 2
b=ts.resample('D').mean()
print('<두번째그룹>')
print(b)

# 3
c=ts.resample('W-MON').mean()
print('<세번째그룹>')
print(c)

# 4
d=ts.resample('MS').mean()
print('<네번째그룹>')
print(d)
```

```python
# open API 는 접어두고 그냥이라도 해보자 - 김진님거 참고

import numpy as np
dates = pd.date_range('1/1/2020', periods=100, freq='W-SAT')
print(dates, type(dates))
```

```python
# 외싯 횟수
df = pd.DataFrame(np.random.randn(100,5),
                 index=dates,
                 columns=['버거','떡볶이','한식','마라','샐러드'])
df.index.name = "음식 먹은 날"
display(df)
```

```python
# 연도별 집계
e_y = df.resample('Y').count()
display(e_y)
# 월별 집계
e_m = df.resample('M').count()
display(e_m)
# 일별 집계
e_d = df.resample('D').count()
display(e_d)
```
